```python
import numpy as np
import os
from PIL import Image
import glob
i = 0
rootDir = '.'
imageDir = 'images'
resizedImageDir = 'ResizedImages'
imageList = []
resizedImageList = []
for dirName, subDirList, fileList in os.walk(rootDir):
    if imageDir in dirName:
        print('Found Directory: %s' % dirName)
        for file in glob.glob(r"{}/*.png".format(imageDir)):
            print('\t%s' %file)
            image = Image.open(file)
            imageList.append(image)
            resized_image = image.resize((200,150))
            resized_image = np.array(resized_image)
            resizedImageList.append(resized_image)
        break
```
